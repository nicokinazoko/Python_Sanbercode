```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
fig,ax = plt.subplots(nrows = 1 , ncols = 3, figsize = (20,10), sharey = True, sharex = True)
ax[0].plot(x,tan , color = 'blue', label = 'tanx' )
ax[1].plot(x,sin, color = 'orange', label = 'sinx' )
ax[2].plot(x,cos, color = 'green', label = 'cosx')
ax[0].set_title('Grafik tanx')
ax[1].set_title('Grafik sinx')
ax[2].set_title('Grafik cosx')
#ax[0].legend(loc = 'center')
#ax[1].legend(loc = 'center')
#ax[2].legend(loc = 'center')
fig.legend(loc = 'upper center')


plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
fig,ax = plt.subplots(figsize=(16,8))
plot1 = ax.inset_axes([-0.1,0.8,0.3,0.3])
plot1.plot(x,y5,color = 'gold')
plot1.plot(x,y,color = 'blue')
plot1.set_title('Grafik CosX^5')

plot2 = ax.inset_axes([-0.1,-0.1,0.3,0.3])
plot2.plot(x,y2,color = 'red')
plot2.plot(x,y,color = 'blue')
plot2.set_title('Grafik CosX^2')

plot3 = ax.inset_axes([0.8,-0.1,0.3,0.3])
plot3.plot(x,y3, color = 'green')
plot3.plot(x,y,color = 'blue')
plot3.set_title('Grafik CosX^3')

plot4 = ax.inset_axes([0.8,0.8,0.3,0.3])
plot4.plot(x,y4, color = 'black')
plot4.plot(x,y,color = 'blue')
plot4.set_title('Grafik CosX^4')

ax.plot(x,y)
ax.set_title('Grafik Cosx')
```
