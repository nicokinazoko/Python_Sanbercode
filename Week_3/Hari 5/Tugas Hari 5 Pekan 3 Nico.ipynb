````python
#code here
fg = sns.load_dataset('diamonds')
def hexbin(x, y, color, **kwargs):
    cmap = sns.light_palette(color, as_cmap=True)
    plt.hexbin(x, y, gridsize=15, cmap=cmap, **kwargs)
    
grid = sns.FacetGrid(df, hue="cut", col="cut",  height=4)
grid.map(hexbin, "carat","price", extent=[0, 3, 0, 20000]); 


# g = sns.FacetGrid(tips, hue="time", col="time", size=4)
# g.map(hexbin, "total_bill", "tip", extent=[0, 3, 0, 200000])
````

# Soal 3: Pair Plot

Ambil 300 data secara acak dari df, kemudian buat pair plotnya dari 300 data tersebut untuk kolom carat, depth, table, dan price sehingga dihasilkan output sesuai expected.

````python
df = df.sample(n=300, random_state=123)
pp = sns.pairplot(df,hue = 'cut', corner = True, vars=["carat","depth","table","price"], markers = ['+','o','s','D','x'])
pp
````

# Soal 4: Joint Plot

Gunakan data yang sama dengan soal 3 untuk menghasilkan dua visualisasi joint plot antara carat dan price nya sesuai expected output.

````python
#code here
import scipy.stats as stats

with sns.axes_style('white'):
    sns.jointplot(df['carat'], df['price'], kind='kde')
    
g = sns.jointplot("carat", "price", data=df, kind='reg');   
g.annotate(stats.pearsonr)
#stat_func=stats.pearsonr
````
