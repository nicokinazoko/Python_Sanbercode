Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
# code here
import matplotlib.pyplot as plt
fig, axis = plt.subplots(figsize=(15,10))
axis.plot(x,cos, color = 'red', label = 'cos x')
axis.plot(x,sin2, marker = 'x', linestyle = ' ', color = 'yellow', label = '2 sin x')
axis.plot(x,cos3,  linestyle = '--', color = 'green', label = '3 cos x')
axis.plot(x,sin, linestyle = '-.', color = 'blue', label = 'sin (x-90)')
axis.legend(loc = 'center')
plt.xlabel('X-Axis Label')
plt.ylabel('Y-Axis Label')

axis.set_title('Grafik Cos Sin')
```
